```python
# Function to Login into Instagram Handle
def instaLogin(username,password):
    try:
        # Find Username and Password textbox, then enter username and password
        wait.until(EC.presence_of_element_located((By.NAME,'username'))).send_keys(username)
        wait.until(EC.presence_of_element_located((By.NAME,'password'))).send_keys(password)

        # Locating Login Button then submit it
        wait.until(EC.presence_of_element_located((By.TAG_NAME,'button'))).submit()
        sleep(1)

        # Locating and clicking "Not now" for saving login info
        wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_ac8f'))).click()
        sleep(1)

        # Locating and clicking "Not Now" for Turn on Notification
        wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_a9-z"]/button[2]'))).click()

    except TimeoutException:
        print("Something Went Wrong!")

# Main Program
# Call instaLogin Function
instaLogin(username,password)
```

```python
# Function to follow the Insta Handle
def Follow():
    try:
        # Locating Follow Button
        button = wait.until(EC.presence_of_element_located((By.TAG_NAME,"button")))

        # Check for Follow status
        if button.text == "Follow":
            # If not Following -> click 
            button.click()
        else:
            # If already following print message
            print("You are already following this Instagram Handle")
        
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
# Function to Unfollow the Insta Handle
def Unfollow():
    try:
        # Locating Follow Button
        button = wait.until(EC.presence_of_element_located((By.TAG_NAME,"button")))

        # Check for Follow status
        if button.text == "Follow":
            # If text on button is Follow means you are already unfollowing the Handle so print message
            print("You are already not following this Instagram Handle")
        else:
            # Click on Following Button
            button.click()
            sleep(2)

            # Now locate and click on Unfollow
            options = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR,"div.x9f619.xjbqb8w.x78zum5.x168nmei.x13lgxp2.x5pf9jr.xo71vjh.x1pi30zi.x1swvt13.x1l90r2v.xyamay9.x1uhb9sk.x1plvlek.xryxfnj.x1c4vz4f.x2lah0s.xdt5ytf.xqjyukv.x1qjc9v5.x1oa3qoh.x1nhvcw1")))
            options[4].click()

    except TimeoutException:
        print("Something Went Wrong!")
```

##### Problem 5 : Like/Unlike posts
Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.<br>
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.


```python
goTOHomepage()
```

```python
def like_top_N_posts(n):
    try:
        # Open the First post
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'div._aabd._aa8k._al3l'))).click()

        count = 1
        while count <= n:
            # Locate the like button and check the text on hovering over it
            span= driver.find_element(By.CSS_SELECTOR,"span._aamw")
            button = span.find_element(By.TAG_NAME,"title")
            button_text = driver.execute_script("return arguments[0].textContent;", button)

            # Check if Hover text of Button is Unlike
            if button_text == "Unlike":
                # If Hover text is Unlike means the post is already liked so print the message
                print(f"Post No. {count} is already liked")
                count += 1
            else:
                # If Hover text is not Unlike -> Click the like button
                wait.until(EC.presence_of_element_located((By.XPATH,'//section/span/div[@role = "button"]'))).click()
                sleep(1)
                count += 1

            wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"div._aaqg._aaqh"))).click()
            sleep(1)

        # After completion return back to Instagram Handle Page
        driver.back()
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
def unlike_top_N_posts(n):
    try:
        # Open the First post
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,'div._aabd._aa8k._al3l'))).click()

        count = 1
        while count <= n:
            # Locate the like button and check the text on hovering over it
            span= driver.find_element(By.CSS_SELECTOR,"span._aamw")
            button = span.find_element(By.TAG_NAME,"title")
            button_text = driver.execute_script("return arguments[0].textContent;", button)

            # Check if Hover text of Button is Unlike
            if button_text == "Unlike":
                # If Hover text is Unlike -> Click to unlike the post
                wait.until(EC.presence_of_element_located((By.XPATH,'//section/span/div[@role = "button"]'))).click()
                sleep(1)
                count += 1
            else:
                # If Hover text not Unlike means the post is already unliked so print the message
                print(f"Post No. {count} is already unliked")
                count += 1

            wait.until(EC.presence_of_element_located((By.CSS_SELECTOR,"div._aaqg._aaqh"))).click()
            sleep(1)

        # After completion return back to Instagram Handle Page
        driver.back()
        
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
# liking the post of dilsefoodie
like_top_N_posts(30)
sleep(2)
```

##### Problem 6 : Extract list of followers
Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.<br>
Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.<br>
Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.


```python
goTOHomepage()
```

```python
def Extract_followers(insta_handle_name):
    Search_and_Open_Handle(insta_handle_name)
    sleep(5)
    
    try:
        # Locating the Followers button and click on it
        wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,"followers"))).click()

        # Locating Followers List Frame
        frame = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aano"]')))

        # Scrolling until first 500 user is located
        for i in range(50):
            sleep(4)
            driver.execute_script('arguments[0].scrollTop=arguments[0].scrollHeight',frame)

        # Getting the handle of all List of user
        handles = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//span[@class="_ap3a _aaco _aacw _aacx _aad7 _aade"]')))

        Users_list = []
        # iterate over handle
        for i in handles[:500]:
            Users_list.append(i.text)
        sleep(1)

        # Locating Cross Button and click on 
        driver.find_element(By.XPATH,'//div[@class="_ac7b _ac7d"]').click()

        return Users_list
    
    except TimeoutException:
        print("Something Went Wrong!")
```

##### Problem 7 : Check the story of ‘coding.ninjas’.
Consider the following Scenarios and print error messages accordingly<br>
If You have already seen the story.<br>
Or The user has no story.<br>
Or View the story if not yet seen.

```python
goTOHomepage()
```

```python
def check_story(insta_handle_name):
    # search and open instagram Handle of given Instagram handle 
    Search_and_Open_Handle(insta_handle_name)
    try:
        # Locating the story element
        story = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aarf _aarg"]')))
        
        #check the Profile photo size to find out story is available or not
        Height = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aarf _aarg"]/canvas'))).get_attribute('height')
        
        if Height == '210':
            print("Viewing the story")
            story.click()
        else:
            print('Already seen the story')
    except:
        # Getting error it means No story is available
        print('No story is available to view')

# Main Program
# Function to Check the story and print message accordingly
insta_handle_name = 'coding.ninjas'
check_story(insta_handle_name)
```

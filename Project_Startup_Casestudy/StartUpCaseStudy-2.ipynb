#### Problem 1: <br>
Your Friend has developed the Product and he wants to establish the product startup and he is searching for a perfect location where getting the investment has a high chance. But due to its financial restriction, he can choose only between three locations - Bangalore, Mumbai, and NCR. As a friend, you want to help your friend deciding the location. NCR include Gurgaon, Noida and New Delhi. Find the location where the most number of funding is done. That means, find the location where startups has received funding maximum number of times. Plot the bar graph between location and number of funding. Take city name "Delhi" as "New Delhi". Check the case-sensitiveness of cities also. That means, at some place instead of "Bangalore", "bangalore" is given. Take city name as "Bangalore". For few startups multiple locations are given, one Indian and one Foreign. Consider the startup if any one of the city lies in given locations.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('startup_funding.csv',encoding = 'utf-8')                              # Convert .csv Data into Data-frame
df.dropna(subset=["CityLocation"],inplace=True)                                         # Remove Null rows for CityLocation column

def separateCity(city):
    return city.split('/')[0].strip()

df['CityLocation'] = df['CityLocation'].apply(separateCity)                             # Separate the city name if two names are given, also correct the City Names for Mumbai and NCR
df['CityLocation'].replace("Delhi","New Delhi",inplace = True)
df['CityLocation'].replace("bangalore","Bangalore",inplace = True)

def separatecommas(amt):
    return float(str(amt).replace(",",""))

df['AmountInUSD'] = df['AmountInUSD'].apply(separatecommas)                             # Separate commas from Amount and convert it into floating data type
df = df.groupby('CityLocation')['AmountInUSD'].count()                                  # Grouping the dataframe by CityLocation and count no of rows of that particular CityLocation
df = df.sort_values(ascending = False)                                                  # Sort in Descending Order

df=df[(df.index =='Bangalore') | (df.index =='Mumbai') | (df.index =='New Delhi') | (df.index =='Gurgaon') | (df.index =='Noida')]      # Filter by Mumbai and NCR region
city = df.index                                                                         # Store the keys in city
totalfunds = df.values                                                                  # Store the values corresponding to the keys in totalfunds

for i in range(len(city)):  
    print(city[i],totalfunds[i])                                                        # Print the Cityname and totalfunds for that city

plt.bar(city ,totalfunds)                                                               # Plotting bar graph with appropriate Xlabels, Ylabels and title
plt.ylabel('NUMBER OF FUNDINGS')
plt.xlabel('LOCATION')
plt.title('LOCATION vs NUMBER OF FUNDINGS')
plt.show()
```

#### Problem 2 : <br>
Even after trying for so many times, your friend’s startup could not find the investment. So you decided to take this matter in your hand and try to find the list of investors who probably can invest in your friend’s startup. Your list will increase the chance of your friend startup getting some initial investment by contacting these investors. Find the top 5 investors who have invested maximum number of times (consider repeat investments in one company also). In a startup, multiple investors might have invested. So consider each investor for that startup. Ignore undisclosed investors.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('startup_funding.csv',encoding = 'utf-8')                          # Convert .csv Data into Data-frame
df=df.dropna(subset=['InvestorsName'])                                              # Remove Null rows for InvestorName column

df = df[df.InvestorsName != 'Undisclosed Investors']                                # Filtering the dataframe for undisclosed Investors
df = df[df.InvestorsName != 'Undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investor']

investors = df.InvestorsName
investors_name = []

for i in investors:
    if i != "":
        temp = i.split(",")
        for j in temp:
            investors_name.append(j.strip())                                        # Storing Investor name data in investor_name list
            
dct = {}
for i in investors_name:
    dct[i] = dct.get(i,0) + 1                                                       # Counting the investor frequerncy using dictionary

dct_keys = sorted(dct,key=dct.get,reverse=True)                                     # Sorting the dictionary in descending order 
c=0
Inv_name=[]
count=[]
for i in range(len(dct_keys)):
    print(dct_keys[i],dct[dct_keys[i]])                                             # Printing the Investor name and frequency of his investment
    Inv_name.append(dct_keys[i])
    count.append(dct[dct_keys[i]])
    c+=1
    if c==5:
        break

plt.pie(count,labels=Inv_name,autopct='%0.2f%%',counterclock=False,startangle=110)  # Plotting a pie graph showing top 5 Investor name and his share of investment
plt.title('Top 5 Investors')
plt.show()
```

#### Problem 3 : <br>
3. After re-analysing the dataset you found out
that some investors have invested in the
same startup at different number of funding
rounds. So before finalising the previous list,
you want to improvise it by finding the top 5
investors who have invested in different
number of startups. This list will be more
helpful than your previous list in finding the
investment for your friend startup. Find the
top 5 investors who have invested maximum
number of times in different companies. That
means, if one investor has invested multiple
times in one startup, count one for that
company. There are many errors in startup
names. Ignore correcting all, just handle the
important ones - Ola, Flipkart, Oyo and
Paytm.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('startup_funding.csv',encoding = 'utf-8')                              # Convert .csv Data into Data-frame
df=df.dropna(subset=['InvestorsName', 'StartupName'])                                   # Remove Null rows for InvestorName and StartupName column

df['StartupName'].replace('Oyorooms','Oyo',inplace = True)                              # Correcting the name of Major Startups                  
df['StartupName'].replace('OyoRooms','Oyo',inplace = True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace = True)
df['StartupName'].replace('OYO Rooms','Oyo',inplace = True)
df['StartupName'].replace('Olacabs','Ola',inplace = True)
df['StartupName'].replace('Ola Cabs','Ola',inplace = True)
df['StartupName'].replace('Olacabs','Ola',inplace = True)
df['StartupName'].replace('Flipkart.com','Flipkart',inplace = True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace = True)
df = df[df.InvestorsName != 'Undisclosed Investors']                                   # Filtering the dataframe for undisclosed Investors 
df = df[df.InvestorsName != 'Undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investor']


investor_name = []
startup_name = []

for index,row in df.iterrows():
    s = row['InvestorsName']
    v = row['StartupName']
    
    i = str(s)
    l = i.split(',')
    
    for j in l:
        if j != '':
            j = j.strip()
            investor_name.append(j)                                                     # Storing the data of Investor name and startup in which he invested in lists
            startup_name.append(v)

df1 = pd.DataFrame({'InvestorsName' : investor_name,'StartupName' : startup_name})      # Storing the list data into new dataframe object
df1 = df1.groupby('InvestorsName')['StartupName'].nunique()                             # Grouping it by InvestorName and filtering the unique elements
df1 = df1.sort_values(ascending = False)                                                # Sorting the data in descending order and storing top 5 rows in dataframe
df1 = df1.head(5)
name = []
count = []
for i in range(5):
    print(df1.index[i],df1.values[i])                                                   # Print the Investor name and no of investments in different companies
    name.append(df1.index[i])
    count.append(df1.values[i])
    
plt.bar(name,count)                                                                     # Plotting a bar graph with proper labels and title
plt.xticks(rotation=40)
plt.xlabel('Investors')
plt.ylabel('No Of Investment')
plt.title('Top 5 Investors')
plt.show()
```

#### Problem 4 : <br>

Find the top 5 investors who have invested in a different number of startups and their investment type is Crowdfunding or Seed Funding.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('startup_funding.csv',encoding = 'utf-8')                                  # Convert .csv Data into Data-frame
df=df.dropna(subset=['InvestorsName', 'StartupName'])                                       # Remove Null rows for InvestorName and StartupName column

df['StartupName'].replace('Oyorooms','Oyo',inplace = True)                                  # Correcting the name of Major Startups 
df['StartupName'].replace('OyoRooms','Oyo',inplace = True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace = True)
df['StartupName'].replace('OYO Rooms','Oyo',inplace = True)
df['StartupName'].replace('Olacabs','Ola',inplace = True)
df['StartupName'].replace('Ola Cabs','Ola',inplace = True)
df['StartupName'].replace('Olacabs','Ola',inplace = True)
df['StartupName'].replace('Flipkart.com','Flipkart',inplace = True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace = True)
df = df[df.InvestorsName != 'Undisclosed Investors']                                        # Filtering the dataframe for undisclosed Investors 
df = df[df.InvestorsName != 'Undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investor']

df['InvestmentType'].replace('PrivateEquity','Private Equity',inplace=True)                 # Correcting the name for type of funding
df['InvestmentType'].replace('SeedFunding','Seed Funding',inplace=True)
df['InvestmentType'].replace('Crowd funding','Crowd Funding',inplace=True)

df = df[(df.InvestmentType == 'Crowd Funding') | (df.InvestmentType == 'Seed Funding')]     # Filtering th data having investment type crowd or seed fuinding

investor_name = []
startup_name = []

for index,row in df.iterrows():
    s = row['InvestorsName']
    v = row['StartupName']
    
    i = str(s)
    l = i.split(',')
    
    for j in l:
        if j != '':
            j = j.strip()
            investor_name.append(j)                                                         # Storing the data of Investor name and startup in which he invested in lists
            startup_name.append(v)

df1 = pd.DataFrame({'InvestorsName' : investor_name,'StartupName' : startup_name})          
df1 = df1.groupby('InvestorsName')['StartupName'].nunique()
df1 = df1.sort_values(ascending = False)
df1 = df1.head(5)
name = []
count = []
for i in range(5):
    print(df1.index[i],df1.values[i])                                                       # Print the Investor name and no of investments in different companies having crowd or seed funding
    name.append(df1.index[i])
    count.append(df1.values[i])
    
plt.bar(name,count)                                                                         # Plot a bar graph showing name of Investor and no of investments
plt.xticks(rotation=60)
plt.xlabel('Investors')
plt.ylabel('No of Investment')
plt.title('Top 5 Investors')
plt.show()
```

#### Problem 5 : <br>
